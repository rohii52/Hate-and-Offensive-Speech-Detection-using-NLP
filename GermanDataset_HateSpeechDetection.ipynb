```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import Cistem
import re
import nltk

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Setting a random seed for reproducibility
np.random.seed(500)

# Initialize the stemmer and stopwords
stemmer = Cistem()
stop_words = stopwords.words('german')

# Defining the headers for the dataset
header_list = ["text_id", "text", "task_1", "task_2"]

# Reading the train and test data
train_path = "/content/drive/MyDrive/Disseration UON/Datasets/german_train.tsv"
test_path = "/content/drive/MyDrive/Disseration UON/Datasets/german_test.tsv"

Corpus_train = pd.read_csv(train_path, encoding='latin-1', sep="\t", names=header_list)
Corpus_test = pd.read_csv(test_path, encoding='latin-1', sep="\t", names=header_list)

# Data Preprocessing and Cleaning
def clean_text(corpus):
    corpus['text'].dropna(inplace=True)
    corpus['text'] = [entry.lower() for entry in corpus['text']]
    corpus['text'] = [re.sub(r'\w+\.\w+', '', entry) for entry in corpus['text']]
    corpus['text'] = [entry.replace(".", " ") for entry in corpus['text']]
    return corpus

Corpus_train = clean_text(Corpus_train)
Corpus_test = clean_text(Corpus_test)

# Tokenization and Stemming
def tokenize_and_stem(corpus):
    for index, entry in enumerate(corpus['text']):
        final_words = []
        words = word_tokenize(entry, language='german')
        for word in words:
            if word not in stop_words and word.isalpha():
                final_words.append(stemmer.stem(word))
        corpus.loc[index, 'text_final'] = ' '.join(final_words)
    return corpus

Corpus_train = tokenize_and_stem(Corpus_train)
Corpus_test = tokenize_and_stem(Corpus_test)

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train = tfidf_vectorizer.fit_transform(Corpus_train['text_final'])
X_test = tfidf_vectorizer.transform(Corpus_test['text_final'])

# Task 1: Logistic Regression on Binary Classification
y_train_task1 = Corpus_train['task_1']
y_test_task1 = Corpus_test['task_1']

# Train/Test split is already done through Corpus_train and Corpus_test, so we proceed to train the model.
log_reg_task1 = LogisticRegression(max_iter=200)
log_reg_task1.fit(X_train, y_train_task1)
pred_task1 = log_reg_task1.predict(X_test)

# Results for Task 1 (Binary Classification)
print("")
print("Logistic Regression for Task 1 (Binary Classification):")
print(f"Accuracy: {accuracy_score(y_test_task1, pred_task1)}")
print(f"Classification Report:\n{classification_report(y_test_task1, pred_task1, digits=4)}")

# Task 2: Logistic Regression on Multiclass Classification
y_train_task2 = Corpus_train['task_2']
y_test_task2 = Corpus_test['task_2']

log_reg_task2 = LogisticRegression(max_iter=200)
log_reg_task2.fit(X_train, y_train_task2)
pred_task2 = log_reg_task2.predict(X_test)

# Results for Task 2 (Multiclass Classification)
print("\nLogistic Regression for Task 2 (Multiclass Classification):")
print(f"Accuracy: {accuracy_score(y_test_task2, pred_task2)}")
print(f"Classification Report:\n{classification_report(y_test_task2, pred_task2, digits=4)}")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import Cistem
import re
import nltk

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Setting a random seed for reproducibility
np.random.seed(500)

# Initialize the stemmer and stopwords
stemmer = Cistem()
stop_words = stopwords.words('german')

# Define the headers for the dataset
header_list = ["text_id", "text", "task_1", "task_2"]

# Reading the train and test data
train_path = "/content/drive/MyDrive/Disseration UON/Datasets/german_train.tsv"
test_path = "/content/drive/MyDrive/Disseration UON/Datasets/german_test.tsv"

Corpus_train = pd.read_csv(train_path, encoding='latin-1', sep="\t", names=header_list)
Corpus_test = pd.read_csv(test_path, encoding='latin-1', sep="\t", names=header_list)

# Data Preprocessing and Cleaning
def clean_text(corpus):
    corpus['text'].dropna(inplace=True)
    corpus['text'] = [entry.lower() for entry in corpus['text']]
    corpus['text'] = [re.sub(r'\w+\.\w+', '', entry) for entry in corpus['text']]
    corpus['text'] = [entry.replace(".", " ") for entry in corpus['text']]
    return corpus

Corpus_train = clean_text(Corpus_train)
Corpus_test = clean_text(Corpus_test)

# Tokenization and Stemming
def tokenize_and_stem(corpus):
    for index, entry in enumerate(corpus['text']):
        final_words = []
        words = word_tokenize(entry, language='german')
        for word in words:
            if word not in stop_words and word.isalpha():
                final_words.append(stemmer.stem(word))
        corpus.loc[index, 'text_final'] = ' '.join(final_words)
    return corpus

Corpus_train = tokenize_and_stem(Corpus_train)
Corpus_test = tokenize_and_stem(Corpus_test)

# Convert text data into a padded sequence format for RNN/BiLSTM
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=5000)
tokenizer.fit_on_texts(Corpus_train['text_final'])

X_train_seq = tokenizer.texts_to_sequences(Corpus_train['text_final'])
X_test_seq = tokenizer.texts_to_sequences(Corpus_test['text_final'])

# Padding sequences to ensure uniform input length
max_length = 180
X_train_padded = tf.keras.preprocessing.sequence.pad_sequences(X_train_seq, maxlen=max_length, padding='post')
X_test_padded = tf.keras.preprocessing.sequence.pad_sequences(X_test_seq, maxlen=max_length, padding='post')

label_encoder_task1 = LabelEncoder()
y_train_task1 = label_encoder_task1.fit_transform(Corpus_train['task_1'])
y_test_task1 = label_encoder_task1.transform(Corpus_test['task_1'])

label_encoder_task2 = LabelEncoder()
y_train_task2 = label_encoder_task2.fit_transform(Corpus_train['task_2'])
y_test_task2 = label_encoder_task2.transform(Corpus_test['task_2'])

# Convert Task 2 labels to categorical (for multiclass classification)
y_train_task2 = tf.keras.utils.to_categorical(y_train_task2, num_classes=4)
y_test_task2 = tf.keras.utils.to_categorical(y_test_task2, num_classes=4)



### RNN and BiLSTM Models for Task 1 (Binary Classification) ###

# Function to build an RNN model for binary classification
def build_rnn_model_binary():
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Embedding(input_dim=5000, output_dim=128, input_length=max_length))
    model.add(tf.keras.layers.SimpleRNN(64, return_sequences=False))
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # Binary classification
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Function to build a BiLSTM model for binary classification
def build_bilstm_model_binary():
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Embedding(input_dim=5000, output_dim=128, input_length=max_length))
    model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=False)))
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # Binary classification
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# RNN model for Task 1
rnn_task1 = build_rnn_model_binary()
rnn_task1.fit(X_train_padded, y_train_task1, epochs=5, batch_size=32, validation_data=(X_test_padded, y_test_task1))

# Evaluate RNN on Task 1
pred_task1_rnn = (rnn_task1.predict(X_test_padded) > 0.5).astype("int32")
print("RNN for Task 1 (Binary Classification):")
print(f"Accuracy: {accuracy_score(y_test_task1, pred_task1_rnn)}")
print(f"Classification Report:\n{classification_report(y_test_task1, pred_task1_rnn, digits=4)}")

# BiLSTM model for Task 1
bilstm_task1 = build_bilstm_model_binary()
bilstm_task1.fit(X_train_padded, y_train_task1, epochs=5, batch_size=32, validation_data=(X_test_padded, y_test_task1))

# Evaluate BiLSTM on Task 1
pred_task1_bilstm = (bilstm_task1.predict(X_test_padded) > 0.5).astype("int32")
print("\nBiLSTM for Task 1 (Binary Classification):")
print(f"Accuracy: {accuracy_score(y_test_task1, pred_task1_bilstm)}")
print(f"Classification Report:\n{classification_report(y_test_task1, pred_task1_bilstm, digits=4)}")
```

# Data Augmentation for Bert


```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
from transformers import BertTokenizer, TFBertForSequenceClassification, AdamWeightDecay

# Load the pre-trained Multilingual BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-uncased')

# Define portions of data to use for training
training_sizes = [0.33, 0.5, 0.75, 1.0]

# Data Preprocessing
header_list = ["text_id", "text", "task_1", "task_2"]

# Reading the train and test data
train_path = "/content/drive/MyDrive/Disseration UON/Datasets/german_train.tsv"
test_path = "/content/drive/MyDrive/Disseration UON/Datasets/german_test.tsv"

Corpus_train = pd.read_csv(train_path, encoding='latin-1', sep="\t", names=header_list)
Corpus_test = pd.read_csv(test_path, encoding='latin-1', sep="\t", names=header_list)

# Encode labels for Task 1 (binary) and Task 2 (multiclass)
label_encoder_task1 = LabelEncoder()
y_train_task1 = label_encoder_task1.fit_transform(Corpus_train['task_1'])
y_test_task1 = label_encoder_task1.transform(Corpus_test['task_1'])

label_encoder_task2 = LabelEncoder()
y_train_task2 = label_encoder_task2.fit_transform(Corpus_train['task_2'])
y_test_task2 = label_encoder_task2.transform(Corpus_test['task_2'])

# Convert Task 2 labels to categorical (for multiclass classification)
y_train_task2 = tf.keras.utils.to_categorical(y_train_task2, num_classes=4)
y_test_task2 = tf.keras.utils.to_categorical(y_test_task2, num_classes=4)

# Tokenize the text using the Multilingual BERT tokenizer
def encode_texts(texts, tokenizer, max_length=128):
    return tokenizer(
        list(texts),
        max_length=max_length,
        padding=True,
        truncation=True,
        return_tensors='tf'
    )

X_train_encodings = encode_texts(Corpus_train['text'], tokenizer)
X_test_encodings = encode_texts(Corpus_test['text'], tokenizer)

# Loop over different training sizes
for size in training_sizes:
    # Get a subset of the training data
    subset_size = int(len(Corpus_train) * size)
    X_train_subset = {key: val[:subset_size] for key, val in X_train_encodings.items()}
    y_train_task1_subset = y_train_task1[:subset_size]
    y_train_task2_subset = y_train_task2[:subset_size]

    print(f"\n--- Training BERT with {int(size * 100)}% of the training data ---")

    ### Multilingual BERT for Task 1 (Binary Classification) ###
    # Load BERT model for binary classification
    bert_model_task1 = TFBertForSequenceClassification.from_pretrained('bert-base-multilingual-uncased', num_labels=1)

    # Compile the model with AdamWeightDecay optimizer
    optimizer = AdamWeightDecay(learning_rate=2e-5, weight_decay_rate=0.01)
    bert_model_task1.compile(optimizer=optimizer,
                             loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                             metrics=['accuracy'])

    # Fine-tune BERT on Task 1
    print(f"Training Task 1 with {int(size * 100)}% of the data")
    bert_model_task1.fit(
        X_train_subset['input_ids'], y_train_task1_subset,
        validation_data=(X_test_encodings['input_ids'], y_test_task1),
        epochs=1, batch_size=16
    )

    # Evaluate BERT for Task 1
    pred_task1_bert = (bert_model_task1.predict(X_test_encodings['input_ids']).logits > 0).astype("int32")
    print(f"Accuracy for Task 1 (Binary) with {int(size * 100)}% of the data: {accuracy_score(y_test_task1, pred_task1_bert)}")
    print(f"Classification Report for Task 1 (Binary) with {int(size * 100)}% of the data:\n{classification_report(y_test_task1, pred_task1_bert, digits=4)}" )

    ### Multilingual BERT for Task 2 (Multiclass Classification) ###
    # Load BERT model for multiclass classification
    bert_model_task2 = TFBertForSequenceClassification.from_pretrained('bert-base-multilingual-cased', num_labels=4)

    # Compile the model using AdamWeightDecay
    bert_model_task2.compile(optimizer=optimizer,
                             loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
                             metrics=['accuracy'])

    # Fine-tune BERT on Task 2
    print(f"Training Task 2 with {int(size * 100)}% of the data")
    bert_model_task2.fit(
        X_train_subset['input_ids'], y_train_task2_subset,
        validation_data=(X_test_encodings['input_ids'], y_test_task2),
        epochs=1, batch_size=16
    )

    # Evaluate BERT for Task 2
    pred_task2_bert = np.argmax(bert_model_task2.predict(X_test_encodings['input_ids']).logits, axis=1)
    y_test_task2_labels = np.argmax(y_test_task2, axis=1)
    print(f"Accuracy for Task 2 (Multiclass) with {int(size * 100)}% of the data: {accuracy_score(y_test_task2_labels, pred_task2_bert)}")
    print(f"Classification Report for Task 2 (Multiclass) with {int(size * 100)}% of the data:\n{classification_report(y_test_task2_labels, pred_task2_bert, digits=4)}")
```
